```python
import pandas as pd
# Change 'fund_master.csv' to whatever your provided master file name is
df_master = pd.read_csv('../data/raw/01_fund_master.csv')

print("Unique AMCs (Fund Houses):", df_master['fund_house'].nunique())
print("Unique Categories:", df_master['category'].nunique())
print("Unique Risk Grades:", df_master['risk_category'].unique())
```

```python
# Change to match your provided historical nav file name
df_nav = pd.read_csv('../data/raw/02_nav_history.csv')

master_codes = set(df_master['amfi_code'].unique())
nav_codes = set(df_nav['amfi_code'].unique())

# Find any codes that are in master but completely missing from the history file
missing_codes = master_codes - nav_codes

print("--- DATA QUALITY SUMMARY ---")
print(f"Total schemes in Master File: {len(master_codes)}")
print(f"Total schemes in NAV History File: {len(nav_codes)}")
if len(missing_codes) == 0:
    print("Perfect Integrity! All master fund codes exist in the history files.")
else:
    print(f"Data Quality Anomaly: {len(missing_codes)} schemes are missing from the history file!")
```
